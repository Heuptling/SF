```python
import pandas as pd
```

```python
flights = pd.read_csv('flights.csv')
```

```python
flights.head()
```

```python
flights = flights.drop(columns=['id','status','scheduled_departure','scheduled_arrival']) #удаляем лишнее
flights.sample(15)
```

```python
flights.columns
```

```python
flights2 = pd.read_csv('flights_NOZ.csv')
flights2.head()
```

```python
flights2 = flights2.drop(columns=['status','scheduled_departure','scheduled_arrival']) #удаляем лишнее
flights2.head()
```

```python
flights = pd.concat([flights, flights2])
flights.info()
```

```python
flights.sample(15)
```

```python
# Чтобы вычислить длительность полётов нужно привести значения времени вылета и прилёта к соответствующему типу.
```

```python
flights.actual_arrival = pd.to_datetime(flights.actual_arrival)
```

```python
flights.actual_departure = pd.to_datetime(flights.actual_departure)
```

```python
flights.info()
```

```python
flights['flight_time'] = flights['actual_arrival']-flights['actual_departure'] #Вычисляем длительность полётов
```

```python
flights.sample(15)
```
